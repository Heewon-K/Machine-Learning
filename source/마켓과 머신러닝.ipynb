#### 도미와 빙어 합치기

* 사이킷런이 기대하는 데이터 형태 = [[길이, 무게], [길이, 무게], ...]

리스트 내포
* fish_data = [[l,w] for l,w in zip(length, weight)]

정답 준비 
* fish_target = [1] * 35 +[0] * 14

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
fish_data = [[l,w] for l,w in zip(length, weight)]
print(fish_data)
```

```python
# 규칙을 찾으려면 정답을 알려주어야 합니다.
# 도미를 찾으려고 함으로 도미는 1, 빙어는 0
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

1. KNN 알고리즘
- k는 작고 홀수 사용
- 최근점 서클 안에 더 많이 있는 카테고리로 예측

- 거리를 재는 방법을 유클리디안 거리 방식
- KNN의 장점 : 이해하기 쉽다
- 단점 : 예측이 느리고 많은 특성을 처리하는 능력이 부족

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)  # fit(2차원)
kn.score(fish_data, fish_target)
```

* 무조건 도미 ==> 주변 가장 가까운 49개를 비교하는데 49개 중 35개는 도미 데이터, 14는 빙어 데이터이므로 무조건 도미라고 예측한다.

### 확인문제

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```
